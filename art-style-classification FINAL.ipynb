```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Conv2D, MaxPooling2D, Flatten, Dense, Dropout, 
    BatchNormalization, GlobalAveragePooling2D, 
    Rescaling, InputLayer, Activation
)
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.initializers import HeNormal
import matplotlib.pyplot as plt
from PIL import Image
from pathlib import Path
from tqdm import tqdm
import shutil

# Configuration
IMAGE_SIZE = 128
BATCH_SIZE = 32
EPOCHS = 30
DATASET_PATH = "/kaggle/input/wikiart"
OPTIMIZED_DATASET_PATH = "/kaggle/working/wikiart_optimized"
OPTIMIZATION_QUALITY = 75

# Set random seeds
np.random.seed(42)
tf.random.set_seed(42)

# Configure GPU
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
    except RuntimeError as e:
        print(e)

def optimize_images(input_path, output_path, quality=75, target_size=(128, 128)):
    """Optimize images by compressing them to reduce file size"""
    print(f"Optimizing images with {quality}% quality...")
    
    # Create output directory
    os.makedirs(output_path, exist_ok=True)
    
    # Get all genre folders
    genres = [d for d in os.listdir(input_path) if os.path.isdir(os.path.join(input_path, d))]
    
    total_processed = 0
    
    for genre in tqdm(genres, desc="Processing genres"):
        genre_input_path = os.path.join(input_path, genre)
        genre_output_path = os.path.join(output_path, genre)
        
        # Create genre output directory
        os.makedirs(genre_output_path, exist_ok=True)
        
        # Get all image files
        image_files = []
        for ext in ['.jpg', '.jpeg', '.png', '.bmp', '.tiff']:
            image_files.extend([f for f in os.listdir(genre_input_path) 
                              if f.lower().endswith(ext.lower())])
        
        # Process each image
        for img_file in tqdm(image_files, desc=f"Processing {genre}", leave=False):
            input_img_path = os.path.join(genre_input_path, img_file)
            output_img_path = os.path.join(genre_output_path, 
                                         Path(img_file).stem + '.jpg')
            
            # Skip if already processed
            if os.path.exists(output_img_path):
                continue
                
            try:
                # Open, resize and compress image
                with Image.open(input_img_path) as img:
                    # Convert to RGB if needed
                    if img.mode != 'RGB':
                        img = img.convert('RGB')
                    
                    # Resize image
                    img = img.resize(target_size, Image.LANCZOS)
                    
                    # Save with compression
                    img.save(output_img_path, 'JPEG', quality=quality, optimize=True)
                    
                total_processed += 1
                
            except Exception as e:
                print(f"Error processing {input_img_path}: {e}")
                continue
    
    print(f"Image optimization complete! Processed {total_processed} images.")
    
    # Calculate size reduction
    try:
        original_size = sum(os.path.getsize(os.path.join(root, file))
                          for root, dirs, files in os.walk(input_path)
                          for file in files)
        optimized_size = sum(os.path.getsize(os.path.join(root, file))
                           for root, dirs, files in os.walk(output_path)
                           for file in files)
        
        reduction = ((original_size - optimized_size) / original_size) * 100
        print(f"Size reduction: {reduction:.1f}%")
        print(f"Original: {original_size / (1024**3):.2f} GB")
        print(f"Optimized: {optimized_size / (1024**3):.2f} GB")
    except:
        print("Could not calculate size reduction")
    
    return output_path

def VGG19_with_batchnorm(input_shape=(128, 128, 3), num_classes=10):
    """VGG19 architecture with BatchNormalization - optimized for P100 GPU"""
    initializer = HeNormal()
    
    model = Sequential()
    model.add(InputLayer(input_shape=input_shape))
    model.add(Rescaling(1./255))  # Normalize input
    
    # Block 1 - Start with more filters for P100
    model.add(Conv2D(64, (3, 3), padding='same', kernel_initializer=initializer))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Conv2D(64, (3, 3), padding='same', kernel_initializer=initializer))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(MaxPooling2D((2, 2)))
    
    # Block 2
    model.add(Conv2D(128, (3, 3), padding='same', kernel_initializer=initializer))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Conv2D(128, (3, 3), padding='same', kernel_initializer=initializer))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(MaxPooling2D((2, 2)))
    
    # Block 3
    model.add(Conv2D(256, (3, 3), padding='same', kernel_initializer=initializer))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Conv2D(256, (3, 3), padding='same', kernel_initializer=initializer))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Conv2D(256, (3, 3), padding='same', kernel_initializer=initializer))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Conv2D(256, (3, 3), padding='same', kernel_initializer=initializer))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(MaxPooling2D((2, 2)))
    
    # Block 4
    model.add(Conv2D(512, (3, 3), padding='same', kernel_initializer=initializer))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Conv2D(512, (3, 3), padding='same', kernel_initializer=initializer))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Conv2D(512, (3, 3), padding='same', kernel_initializer=initializer))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Conv2D(512, (3, 3), padding='same', kernel_initializer=initializer))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(MaxPooling2D((2, 2)))
    
    # Block 5
    model.add(Conv2D(512, (3, 3), padding='same', kernel_initializer=initializer))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Conv2D(512, (3, 3), padding='same', kernel_initializer=initializer))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Conv2D(512, (3, 3), padding='same', kernel_initializer=initializer))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Conv2D(512, (3, 3), padding='same', kernel_initializer=initializer))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(MaxPooling2D((2, 2)))
    
    # Classification head - optimized for P100
    model.add(GlobalAveragePooling2D())
    model.add(Dense(2048, kernel_initializer=initializer))  # Increased for P100
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Dropout(0.5))
    
    model.add(Dense(1024, kernel_initializer=initializer))  # Increased for P100
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Dropout(0.5))
    
    # Final layer uses float32 for numerical stability with mixed precision
    model.add(Dense(num_classes, activation='softmax', dtype='float32'))
    
    return model

def create_data_generators(dataset_path):
    """Create train and validation data generators"""
    train_datagen = ImageDataGenerator(
        rescale=1./255,
        validation_split=0.2,
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest'
    )
    
    val_datagen = ImageDataGenerator(
        rescale=1./255,
        validation_split=0.2
    )
    
    train_generator = train_datagen.flow_from_directory(
        dataset_path,
        target_size=(IMAGE_SIZE, IMAGE_SIZE),
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        subset='training'
    )
    
    val_generator = val_datagen.flow_from_directory(
        dataset_path,
        target_size=(IMAGE_SIZE, IMAGE_SIZE),
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        subset='validation'
    )
    
    return train_generator, val_generator

def plot_history(history):
    """Plot training history"""
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Training')
    plt.plot(history.history['val_accuracy'], label='Validation')
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Training')
    plt.plot(history.history['val_loss'], label='Validation')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    
    plt.tight_layout()
    plt.show()

# Main execution
print("WikiArt Classification with VGG19 + BatchNormalization")
print("=" * 50)

# Optimize images first
print("Step 1: Optimizing images...")
optimized_path = optimize_images(
    DATASET_PATH, 
    OPTIMIZED_DATASET_PATH, 
    quality=OPTIMIZATION_QUALITY,
    target_size=(IMAGE_SIZE, IMAGE_SIZE)
)

# Create data generators using optimized images
print("Step 2: Creating data generators...")
train_gen, val_gen = create_data_generators(optimized_path)

# Get number of classes
num_classes = len(train_gen.class_indices)
print(f"Number of classes: {num_classes}")
print(f"Classes: {list(train_gen.class_indices.keys())}")

# Create model
print("Building VGG19 model...")
model = VGG19_with_batchnorm(num_classes=num_classes)

# Compile model
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("\nModel Summary:")
model.summary()

# Define callbacks
callbacks = [
    EarlyStopping(
        monitor='val_loss',
        patience=5,
        restore_best_weights=True
    ),
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.2,
        patience=5,
        min_lr=1e-7
    ),
    ModelCheckpoint(
        'best_vgg19_model.h5',
        monitor='val_accuracy',
        save_best_only=True,
        mode='max'
    )
]

# Train model
print("Starting training...")
history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=EPOCHS,
    callbacks=callbacks,
    verbose=1
)

# Plot results
plot_history(history)

# Evaluate model
print("Evaluating model...")
val_loss, val_acc = model.evaluate(val_gen, verbose=1)
print(f"Final Validation Accuracy: {val_acc:.4f}")
print(f"Final Validation Loss: {val_loss:.4f}")

# Save final model
model.save('vgg19_wikiart_final.h5')
print("Training complete! Model saved.")

def predict_and_display(model, val_generator, num_images=5):
    """Display sample images with predictions"""
    # Get a batch of validation images
    val_generator.reset()
    images, true_labels = next(val_generator)
    
    # Get predictions
    predictions = model.predict(images[:num_images])
    predicted_classes = np.argmax(predictions, axis=1)
    true_classes = np.argmax(true_labels[:num_images], axis=1)
    
    # Get class names
    class_names = list(val_generator.class_indices.keys())
    
    # Create figure
    plt.figure(figsize=(20, 4))
    
    for i in range(num_images):
        plt.subplot(1, num_images, i + 1)
        
        # Display image
        img = images[i]
        if img.max() <= 1.0:  # If normalized
            img = (img * 255).astype(np.uint8)
        plt.imshow(img)
        
        # Get prediction info
        predicted_class = class_names[predicted_classes[i]]
        true_class = class_names[true_classes[i]]
        confidence = predictions[i][predicted_classes[i]] * 100
        
        # Set title with prediction
        title = f'Predicted: {predicted_class}\n({confidence:.1f}% confidence)'
        if predicted_classes[i] == true_classes[i]:
            plt.title(title, color='green', fontsize=10)
        else:
            title += f'\nActual: {true_class}'
            plt.title(title, color='red', fontsize=10)
        
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()
    
    # Print detailed predictions
    print("\nDetailed Predictions:")
    print("-" * 60)
    for i in range(num_images):
        predicted_class = class_names[predicted_classes[i]]
        true_class = class_names[true_classes[i]]
        confidence = predictions[i][predicted_classes[i]] * 100
        
        status = "✓ CORRECT" if predicted_classes[i] == true_classes[i] else "✗ INCORRECT"
        print(f"Image {i+1}: {predicted_class} ({confidence:.1f}%) - {status}")
        if predicted_classes[i] != true_classes[i]:
            print(f"         True class: {true_class}")
        
        # Show top 3 predictions
        top_3_indices = np.argsort(predictions[i])[-3:][::-1]
        print(f"         Top 3: ", end="")
        for j, idx in enumerate(top_3_indices):
            print(f"{class_names[idx]} ({predictions[i][idx]*100:.1f}%)", end="")
            if j < 2:
                print(", ", end="")
        print("\n")

# Make predictions on sample images
print("\nMaking predictions on sample images...")
predict_and_display(model, val_gen, num_images=5)
```
